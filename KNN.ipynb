# KNN from scrath

```python
from collections import Counter
class KNN:
    def __init__(self,k=5):
        self.k=k
        self.x_train=None
        self.y_train=None
    def fit(self,x_train,y_train):
        self.x_train=x_train
        self.y_train=y_train
    
    def predict(self,x_test):
        predictions=[]
        
        for i in x_test:
            distance=[]
            for x_train,y_train in zip(self.x_train,self.y_train):
                   distance.append(( self.calculate_distance(x_train,i),y_train))
                    
            distance.sort(key=lambda x:x[0])
            neighbors=distance[0:self.k]
            label=[label for _,label in neighbors]
            predictions.append(Counter(label).most_common(1)[0][0])
        return predictions
    def calculate_distance(self,x_train,i):
        return np.linalg.norm(x_train-i)
```
