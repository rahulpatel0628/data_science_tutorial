# Simple Linear Regression without Inbuilt function

```python
class LinearReg():
    def __init__(self):
        self.coef_=None
        self.intercept_=None
    def fit(self,x_train,y_train):
        num=0
        den=0
        for i in range(x_train.shape[0]):
            num=num+ ((x_train[0]-x_train.mean())*(y_train[0]-y_train.mean()))
            den=den+ ((x_train[0]-x_train.mean())**2)
        self.coef_=num/den
        self.intercept_=y_train.mean()-self.coef_*x_train.mean()
    def predict(self,x_test):
        return x_test*self.coef_+self.intercept_
```

```python
l=LinearReg()
```

```python
l.fit(x_train,y_train)
```

```python
y_ans=l.predict(x_test)
```

```python
r2_score(y_test,y_ans)
```
