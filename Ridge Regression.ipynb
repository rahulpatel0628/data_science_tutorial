# Import Libraries

```python
import numpy as np
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Ridge
from sklearn.metrics import r2_score,mean_squared_error
```

# Dataset

```python
data=load_diabetes()
x=data.data
y=data.target
```

# Split into training and testing data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

# Train Model

```python
#Using Linear Regression
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
print("R2 Score:",r2_score(y_test,y_pred))
print("RMSE:",mean_squared_error(y_test,y_pred)**0.5)
```

```python
#Using Ridge Regression
```

```python
r=Ridge(alpha=0.001)
r.fit(x_train,y_train)
```

```python
new_y_pred=r.predict(x_test)
```

```python
print("R2 Score:",r2_score(y_test,new_y_pred))
print("RMSE:",mean_squared_error(y_test,new_y_pred)**0.5)
```

```python
#There is no major difference between linear and ridge rigression on diabetes data
```

# Check underfitting and Overfitting Cases

```python
import matplotlib.pyplot as plt
m=100
x1=5*np.random.randn(m,1)-2
x2=0.7 + x1**2-2*x1+3+np.random.randn(m,1)
plt.plot(x1,x2,"b.")
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
```

```python
def get_preds_ridge(x1,x2,alpha):
    model=Pipeline([('Poly_feats',PolynomialFeatures(degree=16)),
                   ('ridge',Ridge(alpha=alpha))])
    model.fit(x1.x2)
    return model.predict(x1)
```

```python
alphas=[0,20,200]
cs=['r','g','b']
plt.figure(figsize=(10,6))
plt.plot(x1,x2,'b+',label='Datapoints')
for alpha,c in zip(alphas,cs):
    preds=get_preds_ridge(x1,x2,alpha)
    plt.plot(sorted(x1[:,0]),preds[np.argsort(x1[:,0])],c,label='Alpha:{}'.format(alpha)])
plt.legend()
plt.show()
```
